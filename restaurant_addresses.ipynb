######  Notes
1. The project is to extract out street addresses of all restaurants in a specific area of a city.

2. I have used the FourSquare Data platform to extract out restaurant addresses using city, latitude and longitude as API query parameters.

3. I have developed a class named 'restaurant_addresses' which gets initialised by attributes such as city, latitude and longitude.

4. When attributes are provided, the 'address' method(function) in the Class gives us the list of 100 restaurants with the street addresses.


```python
class restaurant_addresses:
    def __init__(self, city, latitude, longitude):
        self.city=city
        self.latitude=latitude
        self.longitude=longitude
    def address(self):
        self.ll=str(self.latitude)+","+str(self.longitude)
        url = 'https://api.foursquare.com/v2/venues/explore'

        params = dict(
        client_id='CVSV4TCP52MVGX0UKFJXETXQATJSUAHBPC35V2IEJSPZ2OWT',
        client_secret='XSRDSAEWQYQUCBPF2WAFTFXS24U0XRKCAS0A0GKDPPLYWWHV',
        v='20180323',
        
        ll=self.ll,
        query='restaurant',
        limit=200
        )
        resp = requests.get(url=url, params=params)
        data = json.loads(resp.text)
        response=data['response']['groups'][0]['items']
        scraped=pd.DataFrame([])
        for i in range(len(response)):
            name=response[i]['venue']['name']

            address=response[i]['venue']['location']['formattedAddress'][0]
            venue_id=response[i]['venue']['id']
            dic=[{"Name":name,"venue_id":venue_id,"Street_Address":address}]
            scraped=scraped.append(dic)
        dataframe=scraped.reset_index(drop=True)
        return dataframe

```
